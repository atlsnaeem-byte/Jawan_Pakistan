# 🟡 Tuple Practice Questions

### Q1. Create a tuple of 5 fruits and print the third fruit.

```python
fruits = ("apple", "banana", "mango", "orange", "grape")
print(fruits[2])
```

### Q2. Count how many times `2` appears in the tuple.

```python
numbers = (1, 2, 3, 2, 4, 2, 5)
print(numbers.count(2))
```

### Q3. Create a tuple of 3 student names and unpack them.

```python
students = ("Ali", "Sara", "Ahmed")
s1, s2, s3 = students
print(s1, s2, s3)
```

### Q4. Find the index of `"Python"` in the tuple.

```python
languages = ("Java", "C++", "Python", "SQL")
print(languages.index("Python"))
```

### Q5. Slice the tuple and print only the middle 3 elements.

```python
nums = (10, 20, 30, 40, 50, 60, 70)
print(nums[2:5])
```

# 🟢 Dictionary Practice Questions

### Q1. Create a dictionary of a student with keys: name, age, marks, city.

```python
student = {
    "name": "Ali",
    "age": 20,
    "marks": 85,
    "city": "Karachi"
}
print(student)
```

### Q2. Print only the keys of the dictionary.

```python
d = {"A": 10, "B": 20, "C": 30}
print(d.keys())
```

### Q3. Update the age of a student from 20 to 21.

```python
student = {"name": "Ali", "age": 20, "city": "Karachi"}
student["age"] = 21
print(student)
```

### Q4. Create a dictionary of 3 products with price and print one product's price.

```python
products = {
    "Laptop": 1200,
    "Phone": 800,
    "Tablet": 500
}
print(products["Phone"])
```

# 🔵 Set Practice Questions

### Q1. Create a set and remove duplicates from the list.

```python
numbers = [1, 2, 2, 3, 4, 4, 5]
unique = set(numbers)
print(unique)
```

### Q2. Find common elements between A and B.

```python
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}
print(A.intersection(B))
```

### Q3. Add "orange" and "mango" to an existing set of fruits.

```python
fruits = {"apple", "banana"}
fruits.add("orange")
fruits.add("mango")
print(fruits)
```

### Q4. Remove "banana" using both remove() and discard().

```python
fruits1 = {"apple", "banana", "mango"}
fruits1.remove("banana")
print("After remove():", fruits1)

fruits2 = {"apple", "banana", "mango"}
fruits2.discard("banana")
print("After discard():", fruits2)
```

### Q5. Find the union of X and Y.

```python
X = {10, 20, 30}
Y = {30, 40, 50}
print(X.union(Y))
```
