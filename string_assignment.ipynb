# String Operations Assignment

**1. Remove extra spaces from both ends of " Hello World ".**

```python
s = " Hello World "
result = s.strip()
print(result)
```

**2. Remove spaces only from the left side of " Start now".**

```python
s = " Start now"
result = s.lstrip()
print(result)
```

**3. Remove trailing spaces from "Finish line ".**

```python
s = "Finish line "
result = s.rstrip()
print(result)
```

**4. Replace "apple" with "banana" in "I have an apple".**

```python
s = "I have an apple"
result = s.replace("apple", "banana")
print(result)
```

**5. Remove the .jpg suffix from "picture.jpg".**

```python
s = "picture.jpg"
result = s.removesuffix(".jpg")
print(result)
```

**6. Remove the "Mr. " prefix from "Mr. John".**

```python
s = "Mr. John"
result = s.removeprefix("Mr. ")
print(result)
```

**7. How many times does the letter "e" appear in "experience"?**

```python
s = "experience"
result = s.count("e")
print(result)
```

**8. Find the position of "@" in the email "user@example.com".**

```python
s = "user@example.com"
result = s.find("@")
print(result)
```
